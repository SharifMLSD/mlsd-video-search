### Fill variables in the next cell & Run the notebook

```python
search_entry = 'a cat'
video_name = 'ALL'
```

```python
import requests
import base64
import io
from PIL import Image

def string_to_image(base64_string):
    image_data = base64.b64decode(base64_string)
    image_stream = io.BytesIO(image_data)
    return Image.open(image_stream)

response = requests.get(f"https://inference-server-mlsd-video-search.darkube.app/{video_name}?search_entry={search_entry}")
if not response.ok:
    raise Exception(f"Request failed. response: {response.raw}, status code: {response.status_code}, reason: {response.request}")
for record in response.json():
    print(f"score: {record['score']}, video name: {record['video_name']}, second: {record['second']}")
    display(string_to_image(record['image_base64']))
```
